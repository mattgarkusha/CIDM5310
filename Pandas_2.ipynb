```python
list1 = [i for i in range(100)]
list2 = [i for i in range(1, 200, 2)]
sampleDfFromLists = pd.DataFrame({'col1': list1, 'col2': list2})
print(sampleDfFromLists.head())
print(sampleDfFromLists.tail())
```

```python
sampleDfFromLists.index
```

```python
letters = pd.Series(map(chr, range(97, 122)))
letters = pd.DataFrame({'letter': (list(letters) + list(letters * 2) + list(letters * 3) + list(letters * 4))})
letters[21:29]
```
