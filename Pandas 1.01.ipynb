# Import Pandas and Data


```python
import pandas as pd
import matplotlib.pyplot as plt

store1_df = pd.read_csv('sample_data/Store1.csv', delimiter=',')
store2_df = pd.read_csv('sample_data/Store2.csv', delimiter=',')
```

# 1. Which store has greater sales?

Store 1.

```python
store1_total_sales = store1_df['Grocery_sales'] + store1_df['Stationary_sales']
store2_total_sales = store2_df['Grocery_sales'] + store2_df['Stationary_sales']

stores = ['Store 1', 'Store 2']
total_sales = [store1_total_sales.sum(), store2_total_sales.sum()]

plt.bar(stores, total_sales)
plt.xlabel('Stores')
plt.ylabel('Total Sales')
plt.title('Total Sales for Each Store')

for i in range(len(stores)):
    plt.text(stores[i], total_sales[i], str(total_sales[i]), ha='center', va='bottom')

plt.show()
```

# 2. Which store has the highest sales for grocery products?

Store 2.

```python
store1_grocery_sales = store1_df['Grocery_sales'].sum()
store2_grocery_sales = store2_df['Grocery_sales'].sum()

# Create a bar chart to visualize grocery sales
stores = ['Store 1', 'Store 2']
grocery_sales = [store1_grocery_sales, store2_grocery_sales]

plt.bar(stores, grocery_sales)
plt.xlabel('Stores')
plt.ylabel('Grocery Sales')
plt.title('Grocery Sales Comparison')

# Add the grocery sales values to each bar
for i in range(len(stores)):
    plt.text(stores[i], grocery_sales[i], str(grocery_sales[i]), ha='center', va='bottom')

plt.show()
```

# 3. Which sotre has the higest sales for March?

Store 1.

```python
# Filter the data for March sales
store1_march_sales = store1_df[store1_df['Months'] == 'Mar']
store2_march_sales = store2_df[store2_df['Months'] == 'Mar']

# Calculate the total sales for each store in March
store1_total_march_sales = store1_march_sales['Grocery_sales'].sum() + store1_march_sales['Stationary_sales'].sum()
store2_total_march_sales = store2_march_sales['Grocery_sales'].sum() + store2_march_sales['Stationary_sales'].sum()

# Create a bar chart to visualize March sales
stores = ['Store 1', 'Store 2']
march_sales = [store1_total_march_sales, store2_total_march_sales]

plt.bar(stores, march_sales)
plt.xlabel('Stores')
plt.ylabel('March Sales')
plt.title('March Sales Comparison')

# Add the March sales values to each bar
for i in range(len(stores)):
    plt.text(stores[i], march_sales[i], str(march_sales[i]), ha='center', va='bottom')

plt.show()
```
